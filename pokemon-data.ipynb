```python
# Filtering Data

# Rule for () & []âœ…:
#Use [] when talking to the DataFrame: â€œGive me this columnâ€ or â€œFilter me these rows.â€
#Use () inside those [] when you're doing a condition with & or |.

df.loc[df['Type 1'] == 'Fire']

#2 condition Filter
df.loc[(df['HP'] > 50) & (df['Type 1'] == 'Fire')].head(5)
#Seperate the conditions by parantheses 
# OR sign (straight line cant find it on keyboard)
```

```python
#Filtered the dataset based on 3 conditions & stored it into legn_fire:
legn_fire = df.loc[(df['Type 1'] == 'Fire') & (df['Defense'] >= 50) & (df['Legendary'] == True)]

#Saved the dataset as CSV file
legn_fire.to_csv('fire_legendary', index=False)

#Retrieved the CSV file and loaded back 
xx = pd.read_csv('fire_legendary')

xx
```

```python
# If we want to filter out certain words -> we use .str.contains: -> Below i will filter out all pokemon with 'Mega' & save them in a variable

mega_poke = df.loc[df['Name'].str.contains('Mega')].head(5)

mega_poke

#If we want NOT mega we use a ~ in the df.loc function
```

```python
#Regex Filtering - filtering data based on textual patterns
import re

# Shows all rows where the PokÃ©monâ€™s name starts with 'Pi' (any case) and is followed by only lowercase letters
df.loc[df['Name'].str.contains('^pi[a-z]*', regex=True, flags=re.I,)].head(5)

#'^Pi[a-z]*' â€” The regex pattern:

#^ â†’ Anchors the match to the start of the string.

#Pi â†’ The name must start with 'Pi' (re.I = caps is ignored.

#[a-z]* â†’ Followed by zero or more lowercase letters.

#ðŸ”¹ Matches: 'Pikachu', 'Pidgey', 'Piplup', etc.
#ðŸ”¸ Doesn't match: 'Caterpie', 'PINEapple' (unless case ignored
```
